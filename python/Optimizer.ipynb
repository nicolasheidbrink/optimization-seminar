```python
np.random.seed(42)
failure_counter = 0
for _ in range(1000):
    ta = TunnelingAlgorithm(fn.altered_cos, fn.altered_cos_grad, [[-10, 10]], verbose=False, fixed_xm=False)
    x_stars, f_star = ta.apply_algorithm(np.array([np.random.uniform(-10, 10)]))
    if len(x_stars) == 1:
        failure_counter += 1
    elif len(x_stars) not in [1, 2]:
        print(f"Unexpected number of minima found: {len(x_stars)} at points {x_stars} with value {f_star}")
        break 
print(failure_counter)
```
