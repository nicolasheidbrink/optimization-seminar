```python
np.random.seed(42)
failure_counter = 0
for _ in range(1000):
    ta = TunnelingAlgorithm(fn.altered_cos, fn.altered_cos_grad, [[-10, 10]], verbose=False, fixed_xm=False)
    x_stars, f_star = ta.apply_algorithm(np.array([np.random.uniform(-10, 10)]))
    if len(x_stars) == 1:
        failure_counter += 1
    elif len(x_stars) not in [1, 2]:
        print(f"Unexpected number of minima found: {len(x_stars)} at points {x_stars} with value {f_star}")
        break 
print(failure_counter)
```

```python
lenx = 2
i = 0
xx = [np.array([-1])]
while lenx ==2 or xx[0] < 0:
    i += 1
    np.random.seed(i)
    ta = TunnelingAlgorithm(fn.altered_cos, fn.altered_cos_grad, [[-10, 10]], verbose=False, fixed_xm=False)
    xx, f_star = ta.apply_algorithm(np.array([np.random.uniform(-10, 10)]))
    lenx = len(xx)
print(f"The seed {i} causes the algorithm to find only one minimum at {xx} with value {f_star}.")
print("The algorithm will be run again with this seed and verbose output.\n")

np.random.seed(i)
ta = TunnelingAlgorithm(fn.altered_cos, fn.altered_cos_grad, [[-10, 10]], verbose=True, fixed_xm=False)
xx, f_star = ta.apply_algorithm(np.array([np.random.uniform(-10, 10)]))
```
